# Otras Funciones

Los pasos realizados como función por si debe ir en una subrutina (En el Challenge lo realizo separado para explicar mi razonamiento en cada paso)

```python
def cleaning_dataFrame(data_frame):
    data_frame['title'] = data_frame['title'].str.replace('Celular', '')
    data_frame['title'] = data_frame['title'].str.replace('Smartphone', '')
    data_frame['title'] = data_frame['title'].str.replace('Telefono', '')
    data_frame['title'] = data_frame['title'].str.replace('Liberado', '')
    data_frame['title'] = data_frame['title'].str.replace('Libre', '')
    data_frame['title'] = data_frame['title'].str.replace('es', '')
    data_frame['title'] = data_frame['title'].str.replace('Barato', '')
    data_frame['title'] = data_frame['title'].str.strip()
    data_frame['brand'] = data_frame['title'].str.split(' ').str[0]
    data_frame = data_frame[['title', 'brand', 'sales', 'unit_sales']]
    return data_frame
```
