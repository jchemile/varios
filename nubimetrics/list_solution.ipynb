```python
d = {
    "Samsung": ["Samsung", "Sansung", "Galaxy", "Samgung"],
    "Philco": ["Philco"],
    "Kanji": ["Kanji"],
    "Nokia": ["Nokia"],
    "Motorola": ["Motorola", "Moto"],
    "Realme": ["Realme"],
    "Oppo": ["Oppo"],
    "Honor": ["Honor"],
    "Lenovo": ["Lenovo"],
    "Xiaomi": ["Xiaomi"],
    "Energizer": ["Energizer"], 
    "Vodafone": ["Vodafone"], 
    "Bgh": ["Bgh"], 
    "Zte": ["Zte"], 
    "Nextel": ["Nextel"], 
    "Blu": ["Blu"], 
    "Quantum": ["Quantum"], 
    "Micromax": ["Micromax"],                      
    "Cat": ["Cat"],                        
    "Lava": ["Lava"],                     
    "Infinix": ["Infinix"], 
    "Wiko": ["Wiko"], 
    "Tecno": ["Tecno"], 
    "Meizu": ["Meizu"], 
    "Vivo": ["Vivo"], 
    "Asus": ["Asus"], 
    "Oneplus": ["Oneplus"], 
    "Microsoft": ["Microsoft"],                      
    "Huawei": ["Huawei"],                        
    "Sony": ["Sony"], 
    "Lg": ["Lg"], 
    "Panasonic": ["Panasonic"], 
    "Plum": ["Plum"], 
    "Yu": ["Yu"], 
    "Verykool": ["Verykool"], 
    "Blackberry": ["Blackberry"], 
    "Alcatel": ["Alcatel"],                      
    "Apple": ["Iphone"], 
    "Iphone": ["Iphone"],            
    "Htc": ["Htc"], 
    "Acer": ["Acer"], 
    "Google": ["Google"] 
}
```

Realizo la función Mapper con la cuál, a través de la lista puedo tener una manera dinámica y organizada
de organizar las diferentes marcas

```python
def mapper(val):
    for key, values in d.items():
        if any(value in val for value in values):
            return key
    return 'Otros'
```

Utilizo la función mapper en la nueva columna que voy a generar

```python
df['brand'] = df.title.map(mapper)
```

Agrupo por brand realizando una suma para poder ver un listado de los productos más vendidos.

```python
df.groupby('brand').sum().sort_values(by=['unit_sales'], ascending=False)
```
