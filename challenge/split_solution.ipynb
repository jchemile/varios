```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('cellphoneslisting.csv', sep='',index_col='id')
df.head()
```

```python
df['title'] = df['title'].str.replace('Celular', '')
df['title'] = df['title'].str.replace('Smartphone', '')
df['title'] = df['title'].str.replace('Telefono', '')
df['title'] = df['title'].str.replace('Liberado', '')
df['title'] = df['title'].str.replace('Libre', '')
df['title'] = df['title'].str.replace('es', '')
df['title'] = df['title'].str.replace('Barato', '')
```

```python
df['title'] = df['title'].str.strip()
```

```python
df.round()
```

```python
df['brand'] = df['title'].str.split(' ').str[0]
```

```python
df = df[['title', 'brand', 'sales', 'unit_sales']]
```

```python
df.head()
```

```python
df.groupby('brand').sum().sort_values('unit_sales', ascending=False)
```
